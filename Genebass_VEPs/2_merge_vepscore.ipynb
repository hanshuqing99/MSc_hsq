```python
# Completeness:
def completeness(VEP_list,df):
    stat_df = pd.DataFrame()
    row = 0
    for i in range(len(VEP_list)): 
        vep = VEP_list[i]
        count_vep_pred = df[df[vep]!='-'].shape[0]
        stat_df.loc[row,'vep'] = vep
        # 21908 is the number of mutations I put into Emsembl
        stat_df.loc[row,'Variants with prediction'] = str(count_vep_pred)+'/'+str(df.shape[0])
        stat_df.loc[row,'Variants with prediction(%)'] = str("{:.2f}".format(count_vep_pred/df.shape[0]*100))+"%"
        row = row + 1

    return stat_df
```

```python
# Since for genebass analysis, the inputs are nucleotide variants, in order to calculate the completeness of different VEP predictions,
# for VEP from Ensembl, we calculate the proportion of valid predictions over all nucleotide variants:
# proportion for each VEP = number of valid predictions (for nucleotide variants) /all nucleotide variants
# for Other VEPs outside Ensembl: Dynamut2 and VESPA, there inputs are amino acid variants, so their proportions are calculated:
# proportion for each VEP = number of valid predictions (for amino acid variants) /all amino acid variants
stat = completeness(VEP_list,drop_df)
```
